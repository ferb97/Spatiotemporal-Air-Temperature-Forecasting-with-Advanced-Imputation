# Creating dataframe dictionary

```python
import os
import pandas as pd

# Folder path where the CSV files are stored
folder_path = "UP_Datasets" 

# Dictionary to store file names and corresponding dataframes
dataframes_dict = {}

# Iterate over all the CSV files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith(".csv"):
        # Read each CSV file into a dataframe
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path)
        
        # Store the dataframe in the dictionary
        dataframes_dict[file_name] = df

# Print the dictionary keys (file names)
print("File names in the dictionary:")
for file_name in dataframes_dict.keys():
    print(file_name)
```

# Removing csv files that do not have the target column

```python
import pandas as pd

# Array to store names of CSV files that contain the "AT (degree C)" column
files_with_at_column = []

for csv_file in dataframes_dict.keys():
    df = dataframes_dict[csv_file]
    
    # Print the initial shape and number of rows
    print(f"\nFile: {csv_file}")
    print(f"  Shape: {df.shape}")
    print(f"  Total Rows: {df.shape[0]}")

    # Check if the column "AT (degree C)" exists
    if "AT (degree C)" in df.columns:
        # Append the file name to the array
        files_with_at_column.append(csv_file)

    else:
        # If the column "AT (degree C)" does not exist
        print("  Not Applicable")

# After processing, print the list of files that contain the "AT (degree C)" column
print(f"\nFiles containing 'AT (degree C)' column: {len(files_with_at_column)}")
print(files_with_at_column)
```

# Removing columns that are present in less than 80% files

```python
import pandas as pd
import os
from collections import Counter

# Dictionary to store column frequencies
column_frequencies = Counter()

# Read each CSV, print the column information, and track column frequencies
file_columns = {}

print("Column Information Before Modification:")
for file_name in files_with_at_column:
    
    # Read the CSV file
    df = dataframes_dict[file_name]
    
    # Store the columns of the current file
    file_columns[file_name] = df.columns.tolist()
    
    # Update the column frequency counter
    column_frequencies.update(df.columns)
    
    # Print the total column number and column names
    print(f"\nFile: {file_name}")
    print(f"Total Columns: {len(df.columns)}")
    print(f"Column Names: {df.columns.tolist()}")

# Calculate threshold (80% of the total files)
threshold = 0.8 * len(files_with_at_column)

# Identify columns that appear in fewer than 80% of the files (columns to remove)
columns_to_remove = [col for col, count in column_frequencies.items() if count < threshold]

# Print the columns that will be removed
print(f"\nColumns to Remove (appear in less than {threshold} files): {columns_to_remove}")

# Remove the columns in 'columns_to_remove' from each CSV file and modify them
print("\nModifying files and removing columns...")
for file_name in files_with_at_column:
    
    # Read the CSV file
    df = dataframes_dict[file_name]
    
    # Identify the columns to be removed from this particular file
    columns_removed = [col for col in columns_to_remove if col in df.columns]
    
    # Drop columns that are in the 'columns_to_remove' list (if they exist in the file)
    df_modified = df.drop(columns=columns_removed, errors='ignore')
    
    # Save the modified dataframe back to CSV
    dataframes_dict[file_name] = df_modified
    
    # Print the current column information after modification, along with the removed columns
    print(f"\nFile: {file_name}")
    print(f"Removed Columns: {columns_removed}")
    print(f"Total Columns After Modification: {len(df_modified.columns)}")
    print(f"Column Names After Modification: {df_modified.columns.tolist()}")
```

# Check column frequencies for selected csv files

```python
import os
import pandas as pd
from collections import defaultdict

# Dictionary to store the frequency of each column and the files where they are not present
column_frequency = defaultdict(int)
missing_columns = defaultdict(list)

# Iterate over each file in the array
for csv_file in files_with_at_column:
    
    # Read the CSV file into a DataFrame
    df = dataframes_dict[csv_file]
    
    # Get the set of columns in this CSV
    columns_in_file = set(df.columns)
    
    # Update the frequency of each column found in this file
    for column in columns_in_file:
        column_frequency[column] += 1
    
    # Check for columns that are missing in this file (i.e., present in other files but not in this one)
    for column in column_frequency:
        if column not in columns_in_file:
            missing_columns[column].append(csv_file)

# Print the results
print("Column Frequency and Missing Files Information:")
for column, frequency in column_frequency.items():
    missing_files = missing_columns.get(column, [])
    missing_files_str = ", ".join(missing_files) if missing_files else "None"
    print(f"{column}: {frequency}, Not present-> {missing_files_str}")
```

# Remove csv files that do not have data before 2022-02-16

```python
import pandas as pd
import os

# Array to hold CSV files with last entry date before the year 2023
files_before_2023 = []
start_date = pd.Timestamp('2022-02-16')

# Loop through each file and print the first and last entry of "From Date" column and number of rows
for file_name in files_with_at_column[:]:  # Iterate over a copy of the list
    
    # Read the CSV file
    df = dataframes_dict[file_name]
    
    # Check if the "From Date" column exists
    if 'From Date' in df.columns:
        # Get the number of rows in the file
        num_rows = len(df)
        
        # Convert the "From Date" column to datetime format
        df['From Date'] = pd.to_datetime(df['From Date'], errors='coerce')
        
        # Get the first and last entry in the "From Date" column
        first_entry = df['From Date'].iloc[0] if num_rows > 0 else "No Data"
        last_entry = df['From Date'].iloc[-1] if num_rows > 0 else "No Data"
        print(f"\nFile: {file_name}")
        print(f"First 'From Date' entry: {first_entry}")
        
        # Check if the last entry is before 2023
        if isinstance(first_entry, pd.Timestamp) and first_entry > start_date:
            files_before_2023.append(file_name)
            files_with_at_column.remove(file_name)  # Remove the file from the list
    else:
        print(f"\nFile: {file_name} does not contain 'From Date' column.")

# Print CSV files with last entry date before 2023
print(f"\nCSV files with first entry date after {start_date}:")
for file_name in files_before_2023:
    print(file_name)
```

# Remove rows before 2022-02-16

```python
import pandas as pd
import os

# Define the cutoff date (16 February, 2022)
cutoff_date = pd.Timestamp(year=2022, month=2, day=16)

# Iterate over each file in the folder
for file_name in files_with_at_column:
    
    # Read the CSV file
    df = dataframes_dict[file_name]
    
    # Check if the "From Date" column exists
    if 'From Date' in df.columns:
        # Convert the "From Date" column to datetime format
        df['From Date'] = pd.to_datetime(df['From Date'], errors='coerce')
        
        # Get the number of rows before filtering
        num_rows_before = len(df)
        
        # Filter out rows with "From Date" earlier than the cutoff date
        df_filtered = df[df['From Date'] >= cutoff_date]
        
        # Get the number of rows after filtering
        num_rows_after = len(df_filtered)
        
        # Calculate the number of rows removed
        num_rows_removed = num_rows_before - num_rows_after
        
        # Print the results for the current file
        print(f"\nFile: {file_name}")
        print(f"Rows before filtering: {num_rows_before}")
        print(f"Rows removed: {num_rows_removed}")
        print(f"Rows remaining: {num_rows_after}")
        
        # Optionally, you can save the filtered DataFrame back to a CSV file
        dataframes_dict[file_name] = df_filtered
    else:
        print(f"\nFile: {file_name} does not contain 'From Date' column.")
```

# Check the target column null values and remove csv files

```python
import pandas as pd

# Initialize lists to keep track of files
removed_files = []

target_null_threshold = 30  # Threshold for null percentage in 'AT (degree C)' column

# Iterate over each file and calculate the null percentage for 'AT (degree C)' column
for file in files_with_at_column:
    
    # Load the CSV file
    df = dataframes_dict[file]
    
    # Calculate the percentage of null values in 'AT (degree C)' column
    if 'AT (degree C)' in df.columns:
        null_percentage = df['AT (degree C)'].isnull().mean() * 100
        print(f"File: {file} | Null Percentage in 'AT (degree C)': {null_percentage:.2f}%")
        
        # Remove files with more than 30% null values
        if null_percentage > target_null_threshold:
            removed_files.append(file)
    else:
        print(f"File: {file} | 'AT (degree C)' column not found")

# Print the number and names of the removed files
print(f"Number of files removed: {len(removed_files)}")
print("Removed files:", removed_files)

for file in removed_files:
    files_with_at_column.remove(file)

# Print the number and names of the remaining files
print(f"Number of files remaining: {len(files_with_at_column)}")
print("Remaining files:", files_with_at_column)
```

# Column Analysis for null values

```python
import pandas as pd
import os

# Dictionary to store details about each column across all files
column_analysis = {}

# Define the null percentage threshold (20%) and the 70% file threshold
null_threshold = 0.2
file_threshold = 0.7 * len(files_with_at_column)

# List to store columns that have less than 70% of files with < 20% null data
columns_with_less_than_70_percent_files_below_20_null = []

# Loop through each file to gather column data and null percentages
for file_name in files_with_at_column: 
    # Read the CSV file
    df = dataframes_dict[file_name]
    
    # Get the percentage of null values for each column in this file
    null_percentage = df.isnull().mean()
    
    # For each column, track its status across files
    for col in df.columns:
        if col not in column_analysis:
            column_analysis[col] = {
                'files_containing_column': [],
                'files_less_than_20_null': [],
                'files_more_than_20_null': [],
                'files_not_containing_column': []
            }
        
        # Mark that the file contains the column
        column_analysis[col]['files_containing_column'].append(file_name)
        
        # Classify the file based on null percentage for this column
        if null_percentage[col] < null_threshold:
            column_analysis[col]['files_less_than_20_null'].append(file_name)
        else:
            column_analysis[col]['files_more_than_20_null'].append(file_name)

# Identify which files do not contain certain columns
all_columns = set(col for file_name in files_with_at_column for col in dataframes_dict[file_name].columns)

for col in all_columns:
    for file_name in files_with_at_column:
        df = dataframes_dict[file_name]
        if col not in df.columns:
            column_analysis[col]['files_not_containing_column'].append(file_name)

# Check for columns that have less than 70% of files with < 20% null data
for col, data in column_analysis.items():
    less_than_20_null_count = len(data['files_less_than_20_null'])
    
    # If less than 70% of the files have < 20% null data for this column, store the column
    if less_than_20_null_count < file_threshold:
        columns_with_less_than_70_percent_files_below_20_null.append(col)

null_threshold_percentage = null_threshold * 100

# Print the results for each column
for col, data in column_analysis.items():
    print(f"\nColumn: {col}")
    
    # Files that do not contain the column
    print(f"Files not containing the column: {data['files_not_containing_column']}")
    
    # Files where null percentage is less than 20%
    print(f"Number of files with < {null_threshold_percentage}% null data: {len(data['files_less_than_20_null'])}")
    print(f"Files with < {null_threshold_percentage}% null data: {data['files_less_than_20_null']}")
    
    # Files where null percentage is greater than or equal to 20%
    print(f"Number of files with >= {null_threshold_percentage}% null data: {len(data['files_more_than_20_null'])}")
    print(f"Files with >= {null_threshold_percentage}% null data: {data['files_more_than_20_null']}")

# Print the columns that have less than 70% of files with < 20% null data
print(f"\nColumns with less than 70% of files containing < {null_threshold_percentage}% null values:")
for col in columns_with_less_than_70_percent_files_below_20_null:
    print(col)
```

# Check for each column, which files have more than 30% null values

```python
import pandas as pd

# Set to store all unique columns across all files
all_columns = set()

# Iterate through the files to build the superset of all columns
for file in files_with_at_column:
    
    # Load the CSV file
    df = dataframes_dict[file]
    
    # Add all columns from the current file to the all_columns set
    all_columns.update(df.columns)

# Convert the set to a list for ordered iteration
all_columns = list(all_columns)
print(f"All Columns({len(all_columns)}): {all_columns}")

# Dictionaries to store results
missing_columns_dict = {}  # Files missing each column
null_percentage_dict = {}  # Files with more than 30% null data for each column

# Outer loop through each column in the all_columns set
for column in all_columns:
    missing_columns_dict[column] = []
    null_percentage_dict[column] = []
    
    # Inner loop through each file to check for missing columns and null percentages
    for file in files_with_at_column:
        
        # Load the CSV file
        df = dataframes_dict[file]
        
        # Check if the column is missing from the current file
        if column not in df.columns:
            missing_columns_dict[column].append(file)
        else:
            # Calculate the null percentage if the column exists
            null_percentage = df[column].isnull().mean() * 100
            
            # Store the file and null percentage if more than 30% null data
            if null_percentage > 30:
                null_percentage_dict[column].append(f"{file} ({null_percentage:.2f}%)")

# Output results

# Print files missing each column
print("Files missing particular columns:")
for column, files in missing_columns_dict.items():
    if files:
        print(f"Column: {column} | Number of files missing column: {len(files)}")
        print(f"Files: {files}")

# Print files with more than 30% null data in each column
print("\nFiles with more than 30% null data in particular columns:")
for column, files in null_percentage_dict.items():
    if files:
        print(f"Column: {column} | Number of files with more than 30% null data: {len(files)}")
        print(f"Files: {files}")
```
